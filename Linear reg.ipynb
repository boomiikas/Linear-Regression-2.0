Electricity bill prediction

```python
df=pd.read_excel('datasets.xlsx',sheet_name="Electricity_Bill")
df.columns
```

```python
X=df[["Units_Consumed"]]
y=df['Bill_Amount']
model=LinearRegression() 
model.fit(X,y)
print("Rate per unit learned:",round(model.coef_[0],2) ) #slope
print("Fixed charge (intercept):", round(model.intercept_,2))
units_test = 1000
predicted_bill = model.predict([[units_test]])[0]
print("Predicted bill:",predicted_bill)
```

```python
plt.scatter(X,y,label="Actual data")
y_predict=model.predict(X)
plt.plot(X,y_predict,color="red",label="Regression line")
plt.xlabel("Units consumed")
plt.ylabel("Bill amount")
plt.legend()
plt.show()
```
